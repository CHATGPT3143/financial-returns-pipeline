```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns





TICKER = ['AAPL','BTC-USD','MSFT','NVDA','TSLA','SPY']
start  = '2017-01-01'
end = '2023-12-31'


data = yf.download(TICKER,start=start,end=end).dropna()
data

# Savving raw data to csv (For ease just manipulate the csv)


data
```

```python
close = data.loc[:,'Close'].copy()

# simple returns
close['AAPL_return'] = close['AAPL'].pct_change().dropna()
close['BTC-USD_return'] = close['BTC-USD'].pct_change().dropna()
close['MSFT_return'] = close['MSFT'].pct_change().dropna()
close['NVDA_return'] = close['NVDA'].pct_change().dropna()
close['TSLA_return'] = close['TSLA'].pct_change().dropna()
close['SPY_return'] = close['SPY'].pct_change().dropna()

close
```

```python
#log returns
close['AAPL_log_return'] = np.log(close['AAPL']/close['AAPL'].shift(1))
close['BTC-USD_log_return'] = np.log(close['BTC-USD']/close['BTC-USD'].shift(1))
close['MSFT_log_return'] = np.log(close['MSFT']/close['MSFT'].shift(1))
close['NVDA_log_return'] = np.log(close['NVDA']/close['NVDA'].shift(1))
close['TSLA_log_return'] = np.log(close['TSLA']/close['TSLA'].shift(1))
close['SPY_log_return'] = np.log(close['SPY']/close['SPY'].shift(1))

close.to_csv('master_data')

```

```python
# simple vs log returns
# Iterate through the list of ticker symbols
for ticker in TICKER:
  plt.figure(figsize=(12,6))
  # Access the columns in the close DataFrame using the ticker symbol
  plt.plot(close[f'{ticker}_return'],label=f'{ticker}_return')
  plt.plot(close[f'{ticker}_log_return'],label=f'{ticker}_log_return')
  plt.legend()
  plt.show()
```

```python
# 21 day_rolliing volatility
close['AAPL_volatility'] = close['AAPL_return'].rolling(window=21).std()
close['BTC-USD_volatility'] = close['BTC-USD_return'].rolling(window=21).std()
close['MSFT_volatility'] = close['MSFT_return'].rolling(window=21).std()
close['NVDA_volatility'] = close['NVDA_return'].rolling(window=21).std()
close['TSLA_volatility'] = close['TSLA_return'].rolling(window=21).std()
close['SPY_volatility'] = close['SPY_return'].rolling(window=21).std()

close = close.dropna()
close
```

```python
close1 = close.loc[:,['AAPL','BTC-USD','MSFT','NVDA','TSLA','SPY']].copy()
close1


#normalising the data

normalised_price = close1/close1.iloc[0]
normalised_price

normalised_price.to_csv('normalised_price.csv')

# plotting

sns.set_style('darkgrid')
normalised_price.plot(figsize=(12,6))
plt.xlabel('Date')
plt.ylabel('Normalised Price')
plt.title('Normalised Price of Stocks')
plt.legend()
plt.show()
```
